```python
import pandas as pd
import numpy as np
from pathlib import Path
import functools
```

```python
path1=Path("/scratch/znazari/PPMI_ver_sep2022/RNA_Seq_data/star_ir2/counts/")
```

```python
bl_files=pd.read_csv(path1/"n_bl.txt",delimiter='\t',header=None )
```

```python
def function_names(fname):
    tokens=fname.split('.')
    return tokens[1] 
```

```python
def merge(a,b):
    a.iloc[:,-2:-1].reset_index(drop=False).merge(b.iloc[:,-2:-1].reset_index(drop=True), left_index=True, right_index=True)
    return a
```

```python
bl_list=[]
for i in range(len(bl_files)):
    bl_list=bl_list+[function_names(bl_files.iloc[i][0])]
    
```

```python
list_bl_files=[]
for i in range(len(bl_files)):
    list_bl_files=list_bl_files+[pd.read_csv(path1/bl_files.iloc[i][0],delimiter='\t', index_col=[0])]
```

```python
list_bl_files[0]
```

```python
for i in range(len(bl_list)):
    k_list=list(list_bl_files[i].columns)
    k_list[-2]=bl_list[i]
    list_bl_files[i].columns=k_list
```

```python
first_two_elements_merge=list_bl_files[0].iloc[:,-2:-1].reset_index(drop=False).merge(list_bl_files[1].iloc[:,-2:-1].reset_index(drop=True), left_index=True, right_index=True)
```

```python
list_bl_files_conv=[first_two_elements_merge]+list_bl_files[2:]
del list_bl_files
def merge_df(a,b):
    return a.reset_index(drop=True).merge(b.iloc[:,-2:-1].reset_index(drop=True),left_index=True, right_index=True)
merged=functools.reduce(merge_df,list_bl_files_conv)
```

```python

del list_bl_files_conv
```
