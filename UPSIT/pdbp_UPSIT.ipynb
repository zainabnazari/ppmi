# UPSIT PDBP Data

```python
from pathlib import Path
import pandas as pd
from datetime import datetime
```

```python
path = Path("/Users/zainabnazari/large_data_ppmi/releases_2023_v4release_1027_clinical")
path1 = Path("/Users/zainabnazari/large_data_ppmi")
```

```python
upsit = pd.read_csv(path/"releases_2023_v4release_1027_clinical_UPSIT.csv")
upsit
```

```python
pdbp_list_upsit = upsit[upsit['participant_id'].str.startswith('PD-')]
data_baseline= pdbp_list_upsit[pdbp_list_upsit['visit_name']=="M0"]
```

```python
with_columns= data_baseline[['participant_id','score_from_booklet_1','score_from_booklet_2','score_from_booklet_3','score_from_booklet_4' ]]
```

```python
with_columns.set_index('participant_id', inplace=True)
with_columns
```

```python
upsit_final_drop = with_columns.dropna(axis=0, how='all')
upsit_final_drop
```

```python
diagnosis_pdbp = pd.read_csv(path/"releases_2023_v4release_1027_clinical_Enrollment.csv",delimiter=',')

filtered_df = diagnosis_pdbp[diagnosis_pdbp['participant_id'].str.startswith('PD')& (diagnosis_pdbp['visit_name'] == 'M0')]
filtered_df
```

```python
diagnosis = filtered_df[['participant_id','study_arm']]
diagnosis
```

```python
# reset the column for the main file
upsit_final= upsit_final_drop.reset_index()
upsit_final
```

```python
merged_df = pd.merge(upsit_final, diagnosis, on='participant_id', how='left')
merged_df
```

```python
merged_df["study_arm"].value_counts()
```

```python
merged_df_filtered = merged_df[merged_df['study_arm'] != 'Disease Control']
merged_df_filtered
```

```python
merged_df_filtered['study_arm'].value_counts()
```

```python
merged_df_filtered.to_csv(path1/"upsit_pdbp_454.tsv" , sep='\t', index=False)
```
