# Distribution of Participants Diagnosis and  Across Different Visits

```python
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
path1 = Path("/home/znazari/data") # where the output data will be saved at the end.
path2=Path("/scratch/znazari/PPMI_ver_sep2022/study_data/Subject_Characteristics/")
```

```python
# reading the file which contains diagnosis
diago=pd.read_csv(path2/"Participant_Status.csv", header=None )
diago1=diago.rename(columns=diago.iloc[0]).drop(diago.index[0]).reset_index(drop=True)
```

```python
# Group by 'COHORT_DEFINITION' and count occurrences
diagnosis_counts = diago1['COHORT_DEFINITION'].value_counts()

# Set a professional color palette with at least one purple
colors = ['grey'] +['tab:purple']+["tab:pink"]+["pink"]+ ["lavenderblush"]+list(plt.cm.Paired(range(len(diagnosis_counts) - 1)))

# Plotting the pie chart
plt.figure(figsize=(12, 6))
plt.subplot(1, 1, 1)  # Creating a subplot for the pie chart
patches, texts, autotexts = plt.pie(
    diagnosis_counts, labels=diagnosis_counts.index, autopct='%1.1f%%', startangle=140, colors=colors,
    wedgeprops=dict(width=0.3, edgecolor='w')  # Add a white border to pie chart wedges
)

# Adding a legend with associated colors
#plt.legend(patches, diagnosis_counts.index, title='Diagnoses', loc='center left', bbox_to_anchor=(1, 0.5), fontsize='small')

plt.title('Distribution of Patient Diagnoses')
plt.tight_layout()  # Ensures that the plots do not overlap
plt.savefig('distribution_patients.png',  transparent=True);
```

```python
visit = pd.read_csv(path1/"visit-data.csv")
```

```python
# Ensure data is numeric for plotting
visit[['PD', 'Prodromal', 'Healthy controls']] = visit[['PD', 'Prodromal', 'Healthy controls']].apply(pd.to_numeric, errors='coerce')

# Define your color palette
colors = ["#800080", "#E0B0FF", "#7F00FF"]  # These are different shades of purple

visit.set_index(['Visit Code', 'Month '])[['PD', 'Prodromal', 'Healthy controls']].plot(kind='bar', stacked=True, color=colors)

#plt.title('Distribution of Individuals')
plt.xlabel('Visit Code, Month')
plt.ylabel('Number of Patients')

plt.tight_layout()
plt.savefig('distribution.png',  transparent=True)
```

```python
# Data
data = {
    'Category': ['30.0 - 39.9', '40.0 - 49.9', '50.0 - 59.9', '60.0 - 69.9', '70.0 - 79.9', '80.0 - 89.9', '90+'],
    'PD': [24, 78, 213, 335, 179, 11, 61],
    'Prodromal': [1, 19, 168, 250, 89, 14, 76],
    'Healthy controls': [10, 21, 65, 85, 40, 8, 8]
}

df = pd.DataFrame(data)

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

categories = df['Category']
pd_bars = plt.bar(categories, df['PD'], label='PD', color='purple')
prodromal_bars = plt.bar(categories, df['Prodromal'], bottom=df['PD'], label='Prodromal', color='#E0B0FF')
healthy_bars = plt.bar(categories, df['Healthy controls'], bottom=df['PD'] + df['Prodromal'], label='Healthy controls', color='#7F00FF')

# Adding labels and title
plt.xlabel('Age Category')
plt.ylabel('Number of Subjects')
plt.title('Distribution of Subjects in Different Age Categories')
plt.legend()


plt.show()
```
