```python
diago=pd.read_csv(path1/"Participant_Status.csv", header=None )
diago1=diago.rename(columns=diago.iloc[0]).drop(diago.index[0]).reset_index(drop=True)
diago1
```

```python
#this is to select the indices of the patients that have these diseases: SNCA, GBA, LRRK2
count_SNCA = (diago1['ENRLSNCA'] != "0")
el_SNCA=count_SNCA[count_SNCA].index.values
el_SNCA
count_GBA = (diago1['ENRLGBA'] != "0")
el_GBA=count_GBA[count_GBA].index.values
el_GBA
count_LRRK2 = (diago1['ENRLLRRK2'] != "0")
el_LRRK2=count_LRRK2[count_LRRK2].index.values
el_LRRK2;
```

```python
rut_=diago1["PATNO"]
```

```python
li_SNCA=rut_[el_SNCA].tolist()
li_GBA=rut_[el_GBA].tolist()
li_LRRK2=rut_[el_LRRK2].tolist()
```

```python
def Union(lst1, lst2, lst3):
    final_list = list(set(lst1) | set(lst2) | set(lst3))
    return final_list
```

```python
patients_eli=Union(li_SNCA, li_GBA,li_LRRK2)
patients_eli;
```

```python
#drop any rows that have "patients_eli", patient IDs in the "PATNO" column
diago1=diago1[diago1.PATNO.isin(patients_eli) == False]
```

```python
diago1
```

```python
diago2=diago1[['PATNO','COHORT_DEFINITION']]
```

```python
diago2
```

```python
age_at_visit_=pd.read_csv(path1/"Age_at_visit.csv", header=None )
age_at_visit1=age_at_visit_.rename(columns=age_at_visit_.iloc[0]).drop(age_at_visit_.index[0]).reset_index(drop=True)
age_at_visit1
```

```python
age_at_visit2=age_at_visit1[age_at_visit1.PATNO.isin(patients_eli) == False]
```

```python
age_at_visit2
```

```python
age_at_visit3=age_at_visit2.loc[age_at_visit2['EVENT_ID'].isin(['BL'])]
age_at_visit3.reset_index(drop=True,inplace=True)
```

```python
del age_at_visit3["EVENT_ID"]
```

```python
path_1=Path("/scratch/znazari/PPMI_ver_sep2022/RNA_Seq_data/star_ir2/counts/")
```

```python
bl_files=pd.read_csv(path_1/"bl.txt", header=None )
```

```python
def function_names(fname):
    tokens=fname.split('.')
    return tokens[1]
```

```python
bl_list=[]
for i in range(len(bl_files)):
    bl_list=bl_list+[function_names(bl_files.iloc[i][0])]
```

```python
big_patients=list(patient_diagonises['PATNO'])
len(big_patients)
```

```python
len(bl_list)
```

```python
contains=[]
not_contains=[]
for i in bl_list:
    if i in big_patients:
        contains+=[i]
    else:
        not_contains+=[i]
```

```python
len(not_contains)
```

```python
len(contains)
```

```python
# selecting rows based on condition 
filtered_PIs = patient_diagonises[patient_diagonises['PATNO'].isin(contains)] 
mfiltered_PI=filtered_PIs.reset_index(drop=True)
mfiltered_PI
```

```python
bl_patients=pd.DataFrame(bl_list)
```

```python
mfiltered_PI['BL_PATNO']=contains
mfiltered_PI.reset_index(drop=True)
```

```python
patno=list(mfiltered_PI['PATNO'])
bl_patno=list(mfiltered_PI['BL_PATNO'])
```

```python
result=[]
for i in bl_patno:
    result+=[patno.index(i)]
```

```python
len(result)
```

```python
index_result=pd.DataFrame(result)
```

```python
mfiltered_PI['Index']=index_result
```

```python
new_data=mfiltered_PI.drop(columns=['PATNO','BL_PATNO'])
```
