## PDBP Data and  patients with at least one parent having PD and Dopaminergic drug use

```python
from pathlib import Path
import pandas as pd
from datetime import datetime
```

```python
# The pathes (Directories) to the data
path1 = Path("/Users/zainabnazari/large_data_ppmi")
path2 = Path("/Users/zainabnazari/large_data_ppmi/releases_2023_v4release_1027_clinical")
```

```python
medical_history = pd.read_csv(path2/"releases_2023_v4release_1027_clinical_PD_Medical_History.csv")
medical_history
```

```python
pdbp_list = medical_history[medical_history['participant_id'].str.startswith('PD-')]
pdbp_list 
```

```python
# List of patients either on levadopa or on dopamine agonist
filtered_ids = pdbp_list.loc[(pdbp_list['on_levodopa'] == 'Yes') | (pdbp_list['on_dopamine_agonist'] == 'Yes'), 'participant_id']
to_be_removed_dopamine = list(filtered_ids)
```

```python
family_history = pd.read_csv(path2/"releases_2023_v4release_1027_clinical_Family_History_PD.csv")
family_history
```

```python
family_list = family_history[family_history['participant_id'].str.startswith('PD-')]
family_list 
```

```python
# Create list of patients where at least one of parents are with PD
filtered_ids2 = family_list.loc[(family_list['biological_mother_with_pd'] == 'Yes') | (family_list['biological_father_with_pd'] == 'Yes'), 'participant_id']
pd_parents_to_be_removed = list(filtered_ids2)
```

```python
# Convert lists to sets and take the union
union_set = set(to_be_removed_dopamine) | set(pd_parents_to_be_removed)

# Convert back to a list if needed
union_list = list(union_set)

len(union_list)
```

```python
main_list = pd.read_csv(path1/"matrix_pdbp_bl_rnaseq_with_diagnosis.tsv", delimiter ='\t')
main_list
```

```python
# dopamine using patients counts

id_column_filtered = main_list[~main_list['participant_id'].isin(to_be_removed_dopamine)]

id_column_filtered = main_list[~main_list['participant_id'].isin(to_be_removed_dopamine)]
id_column_filtered['study_arm'].value_counts()
```

# Removing Part:
We only keep those individuals with parents having no PD so to keep more data for analysis

```python
# cases-controls with PD parents removed
id_column_filtered_no_parents = main_list[~main_list['participant_id'].isin(pd_parents_to_be_removed)]
id_column_filtered_no_parents['study_arm'].value_counts()
```

```python
id_column_filtered_no_parents.set_index('participant_id', inplace = True)
```

```python
id_column_filtered_no_parents.to_csv(path1/"matrix_no_parents_with_pd.tsv", sep='\t', index = True)
```
