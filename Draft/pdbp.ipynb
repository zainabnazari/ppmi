## PDBP Data 

```python
from pathlib import Path
import pandas as pd
```

```python
# The pathes (Directories) to the data
path1 = Path("/Users/zainabnazari/large_data_ppmi")
path2 = Path("/Users/zainabnazari/large_data_ppmi/releases_2023_v4release_1027_clinical")
```

```python
pdbp = pd.read_csv(path1/"releases_2023_v4release_1027_rnaseq-WB-RWTS_subread_feature-counts_matrix.featureCounts.tsv",delimiter='\t')
pdbp
```

```python
# Extract column names starting with "PD" and the "Geneid" column
selected_columns_pdbp = pdbp[[col for col in pdbp.columns if col.startswith('PD') or col == 'Geneid']]
selected_columns_pdbp
```

```python
# Filter columns containing "BLM0T1" in their names : Baseline visit

selected_columns = [col for col in selected_columns_pdbp.columns if 'BLM0T1' in col or col == 'Geneid']

# Create a new DataFrame with the selected columns
selected_BL_pdbp = selected_columns_pdbp[selected_columns]
selected_BL_pdbp
```

```python
selected_BL_pdbp.set_index('Geneid', inplace=True)
selected_BL_pdbp
```

```python
selected_BL_pdbp.to_csv(path1/"matrix_pdbp_bl_rnaseq.tsv" , sep='\t', index=True)
```

## Diagnosis PDBP

```python
diagnosis_pdbp = pd.read_csv(path2/"releases_2023_v4release_1027_clinical_Enrollment.csv",delimiter=',')
diagnosis_pdbp['prodromal_category'].unique()
```

```python
filtered_df = diagnosis_pdbp[diagnosis_pdbp['participant_id'].str.startswith('PD')& (diagnosis_pdbp['visit_name'] == 'M0')]
filtered_df
```

```python
# Remove the '-BLM0T1' suffix from column names of df1
df1_column_names = [name.replace('-BLM0T1', '') for name in selected_BL_pdbp.columns]

# Find common values between modified column names of df1 and 'participant_id' column of df2
common_names = set(df1_column_names) & set(filtered_df['participant_id'])

# If you want to filter df2 based on these common names:
filtered_df2 = filtered_df[filtered_df['participant_id'].isin(common_names)]
```

```python
simple_diag = filtered_df2[['participant_id','study_arm']]
```

```python
simple_diag
```

Nature paper: 804 cases + 442 controls = 1246 total -> PDBP 

```python
simple_diag['study_arm'].unique()
```

```python
new_diagnosis = simple_diag[simple_diag['study_arm'].isin(['Healthy Control', 'PD'])]
new_diagnosis
```

```python
df2 = new_diagnosis
df1 = selected_BL_pdbp
```

```python
# Extract the row names of df2
df2_row_names = df2['participant_id'].tolist()

# Adjust the column names of df1 by removing the "HPC_" at the end
df1_column_names_adjusted = [name[:-7] for name in df1.columns]

# Find the common column names between df1 and df2 after adjustment
common_column_names = set(df1_column_names_adjusted) & set(df2_row_names)

# Filter df1 to keep only the common columns
_df1 = df1[[col for col in df1.columns if col[:-7] in common_column_names]]
_df1
```

```python
len(common_column_names)
```

```python
matrix_ = _df1.T
```

```python
matrix_
```

```python
# Remove the "-BLM0T1" string from the index column of df2
matrix_.index = matrix_.index.str.replace('-BLM0T1', '')

# Merge df1 and df2 on the participant_id column
merged_df = matrix_.merge(new_diagnosis[['participant_id', 'study_arm']], left_index=True, right_on='participant_id', how='left')

# Rename the study_diag column to indicate its source
merged_df.rename(columns={'study_diag': '_study_diag'}, inplace=True)
merged_df
```

```python
merged_df.set_index('participant_id', inplace=True)
merged_df
```

```python
merged_df.to_csv(path1/"matrix_pdbp_bl_rnaseq_with_diagnosis.tsv" , sep='\t', index=True)
```
