```python
import matplotlib.pyplot as plt
import pandas as pd

# Your data
data = {
    'Category': ['30.0 - 39.9', '40.0 - 49.9', '50.0 - 59.9', '60.0 - 69.9', '70.0 - 79.9', '80.0 - 89.9', '90+'],
    'PD': [24, 78, 213, 335, 179, 11, 61],
    'Prodromal': [1, 19, 168, 250, 89, 14, 76],
    'Healthy controls': [10, 21, 65, 85, 40, 8, 8]
}

df = pd.DataFrame(data)

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

categories = df['Category']
pd_bars = plt.bar(categories, df['PD'], label='PD', color='purple')
prodromal_bars = plt.bar(categories, df['Prodromal'], bottom=df['PD'], label='Prodromal', color='#E0B0FF')
healthy_bars = plt.bar(categories, df['Healthy controls'], bottom=df['PD'] + df['Prodromal'], label='Healthy controls', color='#7F00FF')

# Adding labels and title
plt.xlabel('Age Category')
plt.ylabel('Number of Subjects')
plt.title('Distribution of Subjects in Different Age Categories')
plt.legend()


plt.show()
```
