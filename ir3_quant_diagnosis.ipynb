```python
path1=Path("/scratch/znazari/PPMI_ver_sep2022/RNA_Seq_data/my_star_ir3/quant/")
```

```python
bl_files=pd.read_csv(path1/"bl.txt", header=None )
```

```python
def function_names(fname):
    tokens=fname.split('.')
    return tokens[1]
```

```python
bl_list=[]
for i in range(len(bl_files)):
    bl_list=bl_list+[function_names(bl_files.iloc[i][0])]
```

```python
big_patients=list(patient_diagonises['PATNO'])
len(big_patients)
```

```python
len(bl_list)
```

```python
contains=[]
not_contains=[]
for i in bl_list:
    if i in big_patients:
        contains+=[i]
    else:
        not_contains+=[i]
```

```python
len(not_contains)
```

```python
len(contains)
```

```python
# selecting rows based on condition 
filtered_PIs = patient_diagonises[patient_diagonises['PATNO'].isin(contains)] 
mfiltered_PI=filtered_PIs.reset_index(drop=True)
mfiltered_PI
```

```python
bl_patients=pd.DataFrame(bl_list)
```

```python
mfiltered_PI['BL_PATNO']=contains
mfiltered_PI.reset_index(drop=True)
```

```python
patno=list(mfiltered_PI['PATNO'])
bl_patno=list(mfiltered_PI['BL_PATNO'])
```

```python
result=[]
for i in bl_patno:
    result+=[patno.index(i)]
```

```python
len(result)
```

```python
index_result=pd.DataFrame(result)
```

```python
mfiltered_PI['Index']=index_result
```

```python
new_data=mfiltered_PI.drop(columns=['PATNO','BL_PATNO'])
```

```python
sorted_diag=new_data.sort_values(by=['Index']).reset_index(drop=True)
sorted_diag
```
